```python
MM=OlympDF[(OlympDF['Sport']=="Badminton")&(OlympDF['Event']=="singles")&(OlympDF['Medal']=="Gold")& (OlympDF['Event_gender']=="M")]
count=MM["NOC"].value_counts()
print("Two countries have the most medals:")
print(count[:2])
print("The list sorted based on Athlete Name is displayed below:")
MMCount=MM[(MM['NOC']=="INA") | (MM['NOC']=="CHN")]
MMCount.sort_values(by=["Athlete"])


```

```python
recentYears=OlympDF[OlympDF['Edition']>=1984]
topCountries=recentYears['NOC'].value_counts()
print("The Top three countries with most medals are listed below:")
topCountries[:3]
```

```python
MGM=OlympDF[(OlympDF['Event']=="100m")&(OlympDF['Medal']=="Gold")& (OlympDF['Gender']=="Men")]
MGMCEAC=MGM[['City','Edition','Athlete','NOC']]
MGMCEAC.sort_values(by=["Edition"],ascending=False)
```

```python
Countries=OlympDF['NOC'].unique()
LatestEdition=OlympDF[OlympDF['Edition']>=2008]
LatestCountries=LatestEdition['NOC'].unique()

count=0
NoWinCountry=[]
print("The list of Countries that did not win a medal is mentioned below:")
for element in Countries:
  if element not in LatestCountries:
    NoWinCountry.append(element)
    print(element)
    count+=1
print("The Count of counties that did not win a medal in 2008 is : "+ str(count))
```

```python
Event=OlympDF.groupby("Event")["Medal"].count()
print(Event)
```

```python
CountryWins=OlympDF.groupby("NOC")["Medal"].count()

FirstYrMedal=OlympDF.groupby("NOC")["Edition"].min()
RecentYrMedal=OlympDF.groupby("NOC")["Edition"].max()

CountryWinAnalysis=pd.DataFrame({'Medals':CountryWins,'First Win':FirstYrMedal,'Recent Win':RecentYrMedal})
CountryWinAnalysis

```

```python
import matplotlib.pyplot as plt
CHNteam=pd.DataFrame(OlympDF[(OlympDF["NOC"]=="CHN")&(OlympDF["Edition"]==2008)])
CHNteamSplit=CHNteam["Medal"].groupby(by=[OlympDF["Gender"]],as_index=True).count()


Obj1=CHNteamSplit.plot.barh(x='Gender',y='Medals',rot=0, stacked=True,title="Chinese Team Medal Count by Gender - 2008 Edition")
print(CHNteamSplit)
x=Obj1.set_xlabel('Number of Medals')
y=Obj1.set_ylabel('Gender')
```

```python
import seaborn as sb
Obj1=CHNteamSplit.reset_index()
Obj2=sb.barplot(x='Gender',y='Medal',data=Obj1,hue="Gender",dodge=False).set(xlabel='Gender',ylabel='Number of Medals',title="Chinese Team Medal Count by Gender - 2008 Edition(Seaborn)")
```
